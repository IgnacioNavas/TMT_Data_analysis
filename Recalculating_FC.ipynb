```python
import pandas as pd
import numpy as np
from openpyxl.utils.datetime import to_excel
```

```python
raw_abundances = pd.read_excel("Experiment/1_hTERT_HME1/Data/All/Renamed.xlsx", sheet_name="raw_abundances_matrix")
raw_abundances
```

## Log2 transformation

```python
columns = raw_abundances.columns
log2_transformed_df = pd.DataFrame(columns=columns)

for row in raw_abundances.itertuples():
    row_list=[]
    index= row[0]
    for element in row:
        if type(element) == float:
            log2 = np.log2(element)
            row_list.append(log2)
        else:
            row_list.append(element)
        # print(row_list[1:])
        # print(log2_transformed_df.columns)
    log2_transformed_df.loc[len(log2_transformed_df)] = row_list[1:]
    
log2_transformed_df.to_excel("Experiment/1_hTERT_HME1/Data/Processed/Handmade_Log2.xlsx", index=False)
log2_transformed_df
```

## Calculating average of the log2 transformed data

```python
log2_transformed_df=log2_transformed_df.fillna(0)

EGF = {'EGF1' : ["EGF1_rep1", 'EGF1_rep2', "EGF1_rep3", "EGF1_rep4"], 
       'EGF2' : ["EGF2_rep1", 'EGF2_rep2', "EGF2_rep3", "EGF2_rep4"],
       'EGF5' : ["EGF5_rep1", 'EGF5_rep2', "EGF5_rep3", "EGF5_rep4"], 
       'EGF10' : ["EGF10_rep1", 'EGF10_rep2', "EGF10_rep3", "EGF10_rep4"], 
       'EGF90' : ["EGF90_rep1", 'EGF90_rep2', "EGF90_rep3", "EGF90_rep4"]}
INS = {'INS1' : ["INS1_rep1", "INS1_rep2", "INS1_rep3", "INS1_rep4"], 
       'INS2' : ["INS2_rep1", "INS2_rep2", "INS2_rep3", "INS2_rep4"], 
       'INS5' : ["INS5_rep1", "INS5_rep2", "INS5_rep3", "INS5_rep4"], 
       'INS10' : ["INS10_rep1", "INS10_rep2", "INS10_rep3", "INS10_rep4"], 
       'INS90' : ["INS90_rep1", "INS90_rep2", "INS90_rep3", "INS90_rep4"]}
EGFnINS = {'EGFnINS1' : ["EGFnINS1_rep1", 'EGFnINS1_rep2', "EGFnINS1_rep3", "EGFnINS1_rep4"], 
           'EGFnINS2' : ["EGFnINS2_rep1", 'EGFnINS2_rep2', "EGFnINS2_rep3", "EGFnINS2_rep4"],
           'EGFnINS5' : ["EGFnINS5_rep1", 'EGFnINS5_rep2', "EGFnINS5_rep3", "EGFnINS5_rep4"], 
           'EGFnINS10' : ["EGFnINS10_rep1", 'EGFnINS10_rep2', "EGFnINS10_rep3", "EGFnINS10_rep4"], 
           'EGFnINS90' : ["EGFnINS90_rep1", 'EGFnINS90_rep2', "EGFnINS90_rep3", "EGFnINS90_rep4"]}
starvation = ["starve_rep1", "starve_rep2", "starve_rep3", "starve_rep4"]
full = ["full_rep1", "full_rep2", "full_rep3", "full_rep4"]

columns = ["protein_ID", "prot_name", "site", 
           "full_log2_avg", "starvation_log2_avg",
           "EGF1_log2_avg", "EGF2_log2_avg", "EGF5_log2_avg", "EGF10_log2_avg", "EGF90_log2_avg", 
           "INS1_log2_avg", "INS2_log2_avg", "INS5_log2_avg", "INS10_log2_avg", "INS90_log2_avg", 
           "EGFnINS1_log2_avg", "EGFnINS2_log_avg", "EGFnINS5_log_avg", "EGFnINS10_log_avg", "EGFnINS90_log_avg",
           "full_log2_std", "starvation_log2_std",
           "EGF1_log2_std", "EGF2_log2_std", "EGF5_log2_std", "EGF10_log2_std", "EGF90_log2_std", 
           "INS1_log2_std", "INS2_log2_std", "INS5_log2_std", "INS10_log2_std", "INS90_log2_std", 
           "EGFnINS1_log2_std", "EGFnINS2_log_std", "EGFnINS5_log_std", "EGFnINS10_log_std", "EGFnINS90_log_std",
           "full_log2_cv", "starvation_log2_cv",
           "EGF1_log2_cv", "EGF2_log2_cv", "EGF5_log2_cv", "EGF10_log2_cv", "EGF90_log2_cv", 
           "INS1_log2_cv", "INS2_log2_cv", "INS5_log2_cv", "INS10_log2_cv", "INS90_log2_cv", 
           "EGFnINS1_log2_cv", "EGFnINS2_log_cv", "EGFnINS5_log_cv", "EGFnINS10_log_cv", "EGFnINS90_log_cv", "n_reps"]

avg_log2_df = pd.DataFrame(columns = columns)

for row in log2_transformed_df.iterrows():
    phospho_info = []
    mean_list =[]
    std_list = []
    cv_list = []
    phospho_info.append(row[1]['protein_ID'])
    phospho_info.append(row[1]['prot_name'])
    phospho_info.append(row[1]['site'])
    
    full_values_list = []
    for replicate in full:
        value = row[1][replicate]
        full_values_list.append(value)
    while 0 in full_values_list:
        full_values_list.remove(0)
    full_values_mean = np.mean(full_values_list)
    mean_list.append(full_values_mean)
    full_values_std = np.std(full_values_list)
    std_list.append(full_values_std)
    full_values_cv = (full_values_std / full_values_mean) * 100
    cv_list.append(full_values_cv)
    n_reps = [len(full_values_list)]
        
    starvation_values_list = []
    for replicate in starvation:
        value = row[1][replicate]
        starvation_values_list.append(value)
    while 0 in starvation_values_list:
        starvation_values_list.remove(0)
    starvation_values_mean = np.mean(starvation_values_list)
    mean_list.append(starvation_values_mean)
    starvation_values_std = np.std(starvation_values_list)
    std_list.append(starvation_values_std)
    starvation_values_cv = (starvation_values_std / starvation_values_mean) * 100
    cv_list.append(starvation_values_cv)
        
    for key in EGF:
        EGF_values_list = []
        for replicate in EGF[key]:
            value = row[1][replicate]
            EGF_values_list.append(value)
        while 0 in EGF_values_list:
            EGF_values_list.remove(0)
        EGF_values_mean = np.mean(EGF_values_list)
        mean_list.append(EGF_values_mean)
        EGF_values_std = np.std(EGF_values_list)
        std_list.append(EGF_values_std)
        EGF_values_cv = (EGF_values_std / EGF_values_mean) * 100
        cv_list.append(EGF_values_cv)
        
    for key in INS:
        INS_values_list = []
        for replicate in INS[key]:
            value = row[1][replicate]
            INS_values_list.append(value)
        while 0 in INS_values_list:
            INS_values_list.remove(0)
        INS_values_mean = np.mean(INS_values_list)
        mean_list.append(INS_values_mean)
        INS_values_std = np.std(INS_values_list)
        std_list.append(INS_values_std)
        INS_values_cv = (INS_values_std / INS_values_mean) * 100
        cv_list.append(INS_values_cv)
        
    for key in EGFnINS:
        EGFnINS_values_list = []
        for replicate in EGFnINS[key]:
            value = row[1][replicate]
            EGFnINS_values_list.append(value)
        while 0 in EGFnINS_values_list:
            EGFnINS_values_list.remove(0)
        EGFnINS_values_mean = np.mean(EGFnINS_values_list)
        mean_list.append(EGFnINS_values_mean)
        EGFnINS_values_std = np.std(EGFnINS_values_list)
        std_list.append(EGFnINS_values_std)
        EGFnINS_values_cv = (EGFnINS_values_std / EGFnINS_values_mean) * 100
        cv_list.append(EGFnINS_values_cv)
        
    row_list = phospho_info + mean_list + std_list + cv_list + n_reps
    
    avg_log2_df.loc[len(avg_log2_df)] = row_list

    # print(row_list, len(row_list))
    
    
avg_log2_df.to_excel("Experiment/1_hTERT_HME1/Data/Processed/Handmade_avg_log2.xlsx", index=False)
avg_log2_df
```

## Calculate the fold change respect the starvation time point

```python
columns_FC = ["protein_ID", "prot_name", "site", 
              "EGF_full", "EGF_starve",
              "EGF1", "EGF2", "EGF5", "EGF10", "EGF90", 
              "INS_full", "INS_starve",
              "INS1", "INS2", "INS5", "INS10", "INS90", 
              "EGFnINS_full", "EGFnINS_starve",
              "EGFnINS1", "EGFnINS2", "EGFnINS5", "EGFnINS10", "EGFnINS90", 
              "n_reps", "sd_all",
              "EGF_full_std", "EGF_starve_std",
              "EGF1_std", "EGF2_std", "EGF5_std", "EGF10_std", "EGF90_std", 
              "INS_full_std", "INS_starve_std",
              "INS1_std", "INS2_std", "INS5_std", "INS10_std", "INS90_std", 
              "EGFnINS_full_std", "EGFnINS_starve_std",
              "EGFnINS1_std", "EGFnINS2_std", "EGFnINS5_std", "EGFnINS10_std", "EGFnINS90_std" ]
# print(len(columns_FC))

log2_FC_df = pd.DataFrame(columns = columns_FC)
for row in avg_log2_df.iterrows():
    phospho_FC = []
    phospho_FC.append(row[1]['protein_ID'])
    phospho_FC.append(row[1]['prot_name'])
    phospho_FC.append(row[1]['site'])
    
    condition_list = ["full_log2_avg", "starvation_log2_avg",
                      "EGF1_log2_avg", "EGF2_log2_avg", "EGF5_log2_avg", "EGF10_log2_avg", "EGF90_log2_avg", 
                      "full_log2_avg", "starvation_log2_avg",
                      "INS1_log2_avg", "INS2_log2_avg", "INS5_log2_avg", "INS10_log2_avg", "INS90_log2_avg", 
                      "full_log2_avg", "starvation_log2_avg",
                      "EGFnINS1_log2_avg", "EGFnINS2_log_avg", "EGFnINS5_log_avg", "EGFnINS10_log_avg", "EGFnINS90_log_avg"]
    
    for condition in condition_list:
        fold_change = row[1][condition] - row[1]['starvation_log2_avg'] 
        phospho_FC.append(fold_change)
    
    phospho_FC.append(row[1]["n_reps"])
    phospho_FC.append("NaN")
    std_list = ["full_log2_std", "starvation_log2_std",
                "EGF1_log2_std", "EGF2_log2_std", "EGF5_log2_std", "EGF10_log2_std", "EGF90_log2_std", 
                "full_log2_std", "starvation_log2_std",
                "INS1_log2_std", "INS2_log2_std", "INS5_log2_std", "INS10_log2_std", "INS90_log2_std", 
                "full_log2_std", "starvation_log2_std",
                "EGFnINS1_log2_std", "EGFnINS2_log_std", "EGFnINS5_log_std", "EGFnINS10_log_std", "EGFnINS90_log_std",]
    for std in std_list:
        phospho_FC.append(row[1][std])
    
    # print(len(phospho_FC))
    log2_FC_df.loc[len(log2_FC_df)] = phospho_FC
    
log2_FC_df.to_excel("Experiment/1_hTERT_HME1/Data/Processed/Handmade_Log2_FC.xlsx", index=False)
log2_FC_df
```

## Calculate the fold change respect the starvation but using the data already processed by the FGCZ

```python
df_fgcz = pd.read_excel("Experiment/1_hTERT_HME1/Data/All/Renamed.xlsx", sheet_name="stats_normalized_wide")

columns_FC = ["protein_ID", "site",  #, "prot_name"
              "EGF_full", "EGF_starve", 
              "EGF1", "EGF2", "EGF5", "EGF10", "EGF90", 
              "INS_full", "INS_starve",
              "INS1", "INS2", "INS5", "INS10", "INS90", 
              "EGFnINS_full", "EGFnINS_starve",
              "EGFnINS1", "EGFnINS2", "EGFnINS5", "EGFnINS10", "EGFnINS90", 
              "n_reps", "sd_all",
              "EGF_full_std", "EGF_starve_std",
              "EGF1_std", "EGF2_std", "EGF5_std", "EGF10_std", "EGF90_std", 
              "INS_full_std", "INS_starve_std",
              "INS1_std", "INS2_std", "INS5_std", "INS10_std", "INS90_std", 
              "EGFnINS_full_std", "EGFnINS_starve_std",
              "EGFnINS1_std", "EGFnINS2_std", "EGFnINS5_std", "EGFnINS10_std", "EGFnINS90_std" ]

log2_FC_df_fgcz = pd.DataFrame(columns = columns_FC)
for row in df_fgcz.iterrows():
    phospho_FC = []
    phospho_FC.append(row[1]['protein_Id'])
    # phospho_FC.append(row[1]['prot_name'])
    phospho_FC.append(row[1]['site'])
    
    condition_list = ["mean_full", "mean_starve",
                      "mean_EGF1", "mean_EGF2", "mean_EGF5", "mean_EGF10", "mean_EGF90", 
                      "mean_full", "mean_starve",
                      "mean_INS1", "mean_INS2", "mean_INS5", "mean_INS10", "mean_INS90", 
                      "mean_full", "mean_starve",
                      "mean_EGFnINS1", "mean_EGFnINS2", "mean_EGFnINS5", "mean_EGFnINS10", "mean_EGFnINS90"]
    
    for condition in condition_list:
        fold_change = row[1][condition] - row[1]['mean_starve'] 
        phospho_FC.append(fold_change)
    
    phospho_FC.append(row[1]["not_na_starve"])
    phospho_FC.append(row[1]["sd_All"])
    std_list = ["sd_full", "sd_starve",
                "sd_EGF1", "sd_EGF2", "sd_EGF5", "sd_EGF10", "sd_EGF90", 
                "sd_full", "sd_starve",
                "sd_INS1", "sd_INS2", "sd_INS5", "sd_INS10", "sd_INS90", 
                "sd_full", "sd_starve",
                "sd_EGFnINS1", "sd_EGFnINS2", "sd_EGFnINS5", "sd_EGFnINS10", "sd_EGFnINS90",]
    for std in std_list:
        phospho_FC.append(row[1][std])
    
    # print(len(phospho_FC))
    log2_FC_df_fgcz.loc[len(log2_FC_df_fgcz)] = phospho_FC
    

prot_name_df = pd.read_excel("Experiment/1_hTERT_HME1/Data/All/Renamed.xlsx", sheet_name="normalized_abundances_matrix")


prot_name_dir = {}
for row in prot_name_df.itertuples():
    protein_ID = row[1]
    prot_name = row[2]
    prot_name_dir[protein_ID] = prot_name

log2_FC_df_fgcz.insert(1, "prot_name", 0)
for row in log2_FC_df_fgcz.itertuples():
    index = row[0]
    id = row[1]
    log2_FC_df_fgcz.at[index, 'prot_name'] = prot_name_dir[id]

log2_FC_df_fgcz.to_excel("Experiment/1_hTERT_HME1/Data/Processed/Handmade_Log2_FC_from_FGZC.xlsx", index=False)
```
