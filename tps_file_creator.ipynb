# TPS file creator

This code will be used to transform our time series data in the right format to be able to run the Temporal Pathway Synthesizer [TPS](https://github.com/koksal/tps.git).

I am going to create separated files for EGF, INS and EGFnINS

```python
import pandas as pd
import numpy as np
from pandas.core.indexes.base import maybe_extract_name
```

## Time series file

1.  Calculate means of the abundances
2. Filter for at least 2 repetitions and all phosphorilations found (use code in Data_filtration_and _processing)
3. Create individual files for each stimulation condition and refactor columns (these files only contain phosphosite and time points means). Save them as **.tsv**



```python
raw_abundances = pd.read_excel("Experiment/1_hTERT_HME1/Data/All/Renamed.xlsx", sheet_name="raw_abundances_matrix")
raw_abundances.fillna(0, inplace=True)

EGF = {'EGF1' : ["EGF1_rep1", 'EGF1_rep2', "EGF1_rep3", "EGF1_rep4"], 
       'EGF2' : ["EGF2_rep1", 'EGF2_rep2', "EGF2_rep3", "EGF2_rep4"],
       'EGF5' : ["EGF5_rep1", 'EGF5_rep2', "EGF5_rep3", "EGF5_rep4"], 
       'EGF10' : ["EGF10_rep1", 'EGF10_rep2', "EGF10_rep3", "EGF10_rep4"], 
       'EGF90' : ["EGF90_rep1", 'EGF90_rep2', "EGF90_rep3", "EGF90_rep4"]}
INS = {'INS1' : ["INS1_rep1", "INS1_rep2", "INS1_rep3", "INS1_rep4"], 
       'INS2' : ["INS2_rep1", "INS2_rep2", "INS2_rep3", "INS2_rep4"], 
       'INS5' : ["INS5_rep1", "INS5_rep2", "INS5_rep3", "INS5_rep4"], 
       'INS10' : ["INS10_rep1", "INS10_rep2", "INS10_rep3", "INS10_rep4"], 
       'INS90' : ["INS90_rep1", "INS90_rep2", "INS90_rep3", "INS90_rep4"]}
EGFnINS = {'EGFnINS1' : ["EGFnINS1_rep1", 'EGFnINS1_rep2', "EGFnINS1_rep3", "EGFnINS1_rep4"], 
           'EGFnINS2' : ["EGFnINS2_rep1", 'EGFnINS2_rep2', "EGFnINS2_rep3", "EGFnINS2_rep4"],
           'EGFnINS5' : ["EGFnINS5_rep1", 'EGFnINS5_rep2', "EGFnINS5_rep3", "EGFnINS5_rep4"], 
           'EGFnINS10' : ["EGFnINS10_rep1", 'EGFnINS10_rep2', "EGFnINS10_rep3", "EGFnINS10_rep4"], 
           'EGFnINS90' : ["EGFnINS90_rep1", 'EGFnINS90_rep2', "EGFnINS90_rep3", "EGFnINS90_rep4"]}
starvation = ["starve_rep1", "starve_rep2", "starve_rep3", "starve_rep4"]
full = ["full_rep1", "full_rep2", "full_rep3", "full_rep4"]

columns = ["protein_ID", "prot_name", "site", 
           "full_mean", "starvation_mean",
           "EGF1_mean", "EGF2_mean", "EGF5_mean", "EGF10_mean", "EGF90_mean", 
           "INS1_mean", "INS2_mean", "INS5_mean", "INS10_mean", "INS90_mean", 
           "EGFnINS1_mean", "EGFnINS2_mean", "EGFnINS5_mean", "EGFnINS10_mean", "EGFnINS90_mean",
           "n_reps"] 
           

mean_df = pd.DataFrame(columns = columns)


for row in raw_abundances.iterrows():
    phospho_info = []
    mean_list =[]
    phospho_info.append(row[1]['protein_ID'])
    phospho_info.append(row[1]['prot_name'])
    phospho_info.append(row[1]['site'])
    
    full_values_list = []
    for replicate in full:
        value = row[1][replicate]
        full_values_list.append(value)
    while 0 in full_values_list:
        full_values_list.remove(0)
    full_values_mean = np.mean(full_values_list)
    mean_list.append(full_values_mean)
    n_reps = [len(full_values_list)]
        
    starvation_values_list = []
    for replicate in starvation:
        value = row[1][replicate]
        starvation_values_list.append(value)
    while 0 in starvation_values_list:
        starvation_values_list.remove(0)
    starvation_values_mean = np.mean(starvation_values_list)
    mean_list.append(starvation_values_mean)
        
    for key in EGF:
        EGF_values_list = []
        for replicate in EGF[key]:
            value = row[1][replicate]
            EGF_values_list.append(value)
        while 0 in EGF_values_list:
            EGF_values_list.remove(0)
        EGF_values_mean = np.mean(EGF_values_list)
        mean_list.append(EGF_values_mean)
        
    for key in INS:
        INS_values_list = []
        for replicate in INS[key]:
            value = row[1][replicate]
            INS_values_list.append(value)
        while 0 in INS_values_list:
            INS_values_list.remove(0)
        INS_values_mean = np.mean(INS_values_list)
        mean_list.append(INS_values_mean)
        
    for key in EGFnINS:
        EGFnINS_values_list = []
        for replicate in EGFnINS[key]:
            value = row[1][replicate]
            EGFnINS_values_list.append(value)
        while 0 in EGFnINS_values_list:
            EGFnINS_values_list.remove(0)
        EGFnINS_values_mean = np.mean(EGFnINS_values_list)
        mean_list.append(EGFnINS_values_mean)
        
    row_list = phospho_info + mean_list + n_reps # std_list + cv_list + n_reps
    # print(len(row_list), len(columns))
    mean_df.loc[len(mean_df)] = row_list
    # print(row_list, len(row_list))
    
# mean_df.to_excel("Experiment/1_hTERT_HME1/Data/Processed/Handmade_abs_mean_nreps.xlsx", index=False)
# mean_df
```

```python
filtered_df = pd.read_excel("Experiment/1_hTERT_HME1/Data/Processed/Handmade_abs_mean_nRep1_allPhosphoFound_True.xlsx")

EGF_columns = ["site", "full_mean", "starvation_mean", "EGF1_mean", "EGF2_mean", "EGF5_mean", "EGF10_mean", "EGF90_mean"]
INS_columns = ["site", "full_mean", "starvation_mean", "INS1_mean", "INS2_mean", "INS5_mean", "INS10_mean", "INS90_mean"]
EGFnINS_columns = ["site", "full_mean", "starvation_mean", "EGFnINS1_mean", "EGFnINS2_mean", "EGFnINS5_mean", "EGFnINS10_mean", "EGFnINS90_mean"]

EGF_time_series = filtered_df[EGF_columns]
EGF_time_series.rename(columns={"full_mean" : "0min", 
                                "starvation_mean" : "120min",
                                "EGF1_mean" : "121min",
                                "EGF2_mean" : "122min",
                                "EGF5_mean" : "125min",
                                "EGF10_mean" : "130min",
                                "EGF90_mean" : "210min"}, inplace=True)
EGF_time_series.to_csv("Experiment/1_hTERT_HME1/Data/Processed/tps_files/EGF_time_series.tsv", sep="\t", index=False)

INS_time_series = filtered_df[INS_columns]
INS_time_series.rename(columns={"full_mean" : "0min", 
                                "starvation_mean" : "120min",
                                "INS1_mean" : "121min",
                                "INS2_mean" : "122min",
                                "INS5_mean" : "125min",
                                "INS10_mean" : "130min",
                                "INS90_mean" : "210min"}, inplace=True)
INS_time_series.to_csv("Experiment/1_hTERT_HME1/Data/Processed/tps_files/INS_time_series.tsv", sep="\t", index=False)

EGFnINS_time_series = filtered_df[EGFnINS_columns]
EGFnINS_time_series.rename(columns={"full_mean" : "0min", 
                                "starvation_mean" : "120min",
                                "EGFnINS1_mean" : "121min",
                                "EGFnINS2_mean" : "122min",
                                "EGFnINS5_mean" : "125min",
                                "EGFnINS10_mean" : "130min",
                                "EGFnINS90_mean" : "210min"}, inplace=True)
EGFnINS_time_series.to_csv("Experiment/1_hTERT_HME1/Data/Processed/tps_files/EGFnINS_time_series.tsv", sep="\t", index=False)

EGF_time_series_no_full = EGF_time_series.drop(["0min"], axis=1)
EGF_time_series_no_full.rename(columns={"120min" : "0min",
                                            "121min" : "1min", 
                                            "122min" : "2min",
                                            "125min" : "5min",
                                            "130min" : "10min",
                                            "210min" : "90min"}, inplace=True)
EGF_time_series_no_full.to_csv("Experiment/1_hTERT_HME1/Data/Processed/tps_files/EGF_time_series_no_full.tsv", sep="\t", index=False)

INS_time_series_no_full = INS_time_series.drop(["0min"], axis=1)
INS_time_series_no_full.rename(columns={"120min" : "0min",
                                            "121min" : "1min", 
                                            "122min" : "2min",
                                            "125min" : "5min",
                                            "130min" : "10min",
                                            "210min" : "90min"}, inplace=True)
INS_time_series_no_full.to_csv("Experiment/1_hTERT_HME1/Data/Processed/tps_files/INS_time_series_no_full.tsv", sep="\t", index=False)

EGFnINS_time_series_no_full = EGFnINS_time_series.drop(["0min"], axis=1)
EGFnINS_time_series_no_full.rename(columns={"120min" : "0min",
                                            "121min" : "1min", 
                                            "122min" : "2min",
                                            "125min" : "5min",
                                            "130min" : "10min",
                                            "210min" : "90min"}, inplace=True)
EGFnINS_time_series_no_full.to_csv("Experiment/1_hTERT_HME1/Data/Processed/tps_files/EGFnINS_time_series_no_full.tsv", sep="\t", index=False)
```
